```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None


class Solution:
    
    def dfs_preorder(self, tree):
        
        # mid -> left -> right
        preorder = []
        stack = [tree]

        while stack:
            node = stack.pop()
            preorder.append(node.val)

            if node.right:
                stack.append(node.right)
            
            if node.left:
                stack.append(node.left)
        
        return preorder


    def dfs_inorder(self, tree):

        # left -> mid -> right
        inorder = []
        stack = []
        cur = tree

        while cur or stack:
            while cur:
                stack.append(cur)
                cur = cur.left
            
            node = stack.pop()
            inorder.append(node.val)
            cur = node.right
        
        return inorder

    def buildtree(self, pre_data, in_data):

        # base case, return None
        if not pre_data or not in_data:
            return None

        root_val = pre_data[0]
        tree = TreeNode(root_val)

        # use inorder
        # in front of root node will be left child 
        # after root node will be right child
        idx = in_data.index(root_val)

        tree.left = self.buildtree(pre_data[1 : idx + 1], in_data[0 : idx])
        tree.right = self.buildtree(pre_data[idx + 1 : ], in_data[idx + 1 : ])

        return tree

    """
    @param root: An object of TreeNode, denote the root of the binary tree.
    This method will be invoked first, you should design your own algorithm 
    to serialize a binary tree which denote by a root node to a string which
    can be easily deserialized by your own "deserialize" method later.
    """
    def serialize(self, root):
        # write your code here

        # base case
        if not root:
            return root
        
        # traversal preorder at first
        preorder = self.dfs_preorder(root)

        # traversal inorder at second
        inorder = self.dfs_inorder(root)

        # if we have two traversal structure, we can built same tree again
        return [preorder, inorder]


    """
    @param data: A string serialized by your serialize method.
    This method will be invoked second, the argument data is what exactly
    you serialized at method "serialize", that means the data is not given by
    system, it's given by your own serialize method. So the format of data is
    designed by yourself, and deserialize it here as you serialize it in 
    "serialize" method.
    """
    def deserialize(self, data):
        # write your code here
        if not data:
            return data
        
        pre_data, in_data = data[0], data[1]
        tree = self.buildtree(pre_data, in_data)

        return tree
        
```
